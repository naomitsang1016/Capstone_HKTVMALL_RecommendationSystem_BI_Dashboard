```python
#Import NLP module
import pycantonese
from pycantonese.word_segmentation import Segmenter
import nltk
from nltk.corpus import stopwords

#Import word processing module
import re
import string
```

```python
#Define English stopwords (nltk stopwords + custom stopwords)
custom_eng_stop_words = ['divclass','div','review','span','spanclass','hktvmall','class','br','<br','l','src','https','nan','lt','href','gt','che','nil','amp','un','bb','ot','hktv','mall','images','img']
eng_stop_words = stopwords.words('english') + custom_eng_stop_words
eng_stop_words
```

```python
#Define Chinese stopwords
custom_chi_stop_words = ['此','為','的','評論','內容','已','用戶','非','公開','嚟','系','但','而且','啦','味','麵','設定','啦','個人','認','今次','其實','圖片','還','雖然','不過','商品','亦','可惜','但是','但','是']
chi_stop_words = pycantonese.stop_words(add=custom_chi_stop_words,remove=['好','唔','唔使','唔係','冇'])
chi_stop_words
```

```python
#Specify a specific linked text when segment
segmenter = Segmenter(allow={"唔好食","唔喜歡","唔鍾意",'好味','好食','抵食','抵買','杯麵','杯面','多次','回購','唔洗','平過','唔算','唔太','唔使','貨品','飲開'})
```

```python
#Define punctuations
eng_punc = string.punctuation
chi_punc = "！？｡。＂＃＄％＆＇（）＊＋，－／：；＜＝＞＠［＼］＾＿｀｛｜｝～｟｠｢｣､、〃》「」『』【】〔〕〖〗〘〙〚〛〜〝〞〟〰〾〿–—‘’‛“”„‟…‧﹏. ..."
punc = eng_punc + chi_punc
punc
```

```python
#Function to identify English text
def find_eng_words(word):
    if re.findall('[a-zA-Z]+',word):
        return re.findall('[a-zA-Z]+',word)[0]
```

```python
#Function to remove emoji
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

```python
#Function to check if there's number in the row value
def has_numbers(inputString):
    return any(char.isdigit() for char in inputString)
```

```python
#Function to remove punctuations
def remove_punctuations(text):
    for punctuation in punc:
        text = text.replace(punctuation, ' ')
    return text
```

```python
#Copy a new dataframe
df_clean_comments = df_clean_less.copy()
df_clean_comments
```

```python
#Steps to clean comments row to show full sentence

#remove emoji
df_clean_comments['Comments_Show'] = df_clean_comments['Comments'].apply(lambda x: remove_emoji(str(x)))

#remove hktvmall reply
df_clean_comments['Comments_Show'] = df_clean_comments['Comments_Show'].apply(lambda x: x.partition('</span>')[0])

#remove rows with nothing
df_clean_comments = df_clean_comments[df_clean_comments.Comments_Show != ""]

#remove rows with pure digits
df_clean_comments = df_clean_comments[df_clean_comments['Comments_Show'].str.isdigit()==False]

#remove rows startwith punctuations (assume = pure punc)
df_clean_comments = df_clean_comments[df_clean_comments['Comments_Show'].str.startswith(punc)==False]

df_clean_comments
```

```python
#Steps to clean comments row for word cloud

#remove emoji
df_clean_comments['Comments_Cloud'] = df_clean_comments['Comments'].apply(lambda x: remove_emoji(str(x)))

#remove hktvmall reply
df_clean_comments['Comments_Cloud'] = df_clean_comments['Comments_Cloud'].apply(lambda x: x.partition('</span>')[0])

#change all english to lowercase
df_clean_comments['Comments_Cloud'] = df_clean_comments['Comments_Cloud'].str.lower()

#remove english stopwords
df_clean_comments['Comments_Cloud'] = df_clean_comments['Comments_Cloud'].apply(lambda x: ' '.join([word for word in nltk.word_tokenize(str(x)) if word.lower() not in stop]))

#remove all digits on rows
df_clean_comments['Comments_Cloud'] = df_clean_comments['Comments_Cloud'].str.replace('\d+', '')

#remove all punc
df_clean_comments['Comments_Cloud'] = df_clean_comments['Comments_Cloud'].apply(remove_punctuations)

#remove rows with nothing
df_clean_comments = df_clean_comments[df_clean_comments.Comments_Cloud != ""]

df_clean_comments
```
