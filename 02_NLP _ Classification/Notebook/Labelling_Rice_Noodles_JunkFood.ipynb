```python
import pandas as pd
import numpy as np
```

```python
df_noodles=pd.read_csv('df_item(noodles).csv')
```

```python
df_noodles['Main_Cat']=['Noodles' for i in df_noodles['Name']]
```

```python
def label_noodles(x):
    if ('烏冬' in x) or ('直條麵' in x):
        return 'Udon noodles'
    elif ('米粉' in x) or ('河粉' in x) or ('米線' in x) or ('冬粉' in x):
        return 'Rice/Cellophane noodles'
    elif ('意粉' in x) or ( 'Pasta' in x) or ('好ｅ食' in x) or ('螺絲粉' in x) or ('達維娜' in x) or ('意大利粉' in x) or ('通心粉' in x) or ('通粉' in x):
        return 'Pasta/Macaroni'
    elif '伊麵' in x:
        return 'Yimian'
    elif '拉麵' in x:
        return 'Ramen'
    elif ('杯麵' in x) or ('合味道' in x):
        return 'Cup noodles'
    elif '粉絲' in x:
        return 'Bean thread'
    elif '香港代理商' in x:
        return 'Others'
    elif ('飯' in x) or ('粥' in x):
        return 'Instant rice/congee'
    elif ('酸辣粉' in x) or ('薯粉' in x) or ('鼎峰園' in x):
        return 'Potato noodles'
    elif ('蒟蒻' in x) or ('魔芋' in x)or ('芋絲' in x):
        return 'Konjac'
    elif '麵'in x:
        return 'Instant noodles'
    else:
        return 'Others'
    
    
    
df_noodles['Sub_Cat']=df_noodles['Name'].apply(lambda x: label_noodles(x))
```

```python
df_noodles['Sub_Cat'].unique()
```

```python
df_noodles.to_csv('df_item(noodles)_updated.csv',index=False)
```

```python
df_noodles['Sub_Cat'].value_counts()
```

```python
df_rice=pd.read_csv('df_item(rice).csv')
```

```python
df_rice['Main_Cat']=['Rice/Oil' for i in df_rice['Name']]
```

```python
def label_rice(x):
    if ('米' in x) :
        return 'Rice'
    
    elif ('油' in x ) :
        return 'Oil'
 
df_rice['Sub_Cat']=df_rice['Name'].apply(lambda x: label_rice(x))
```

```python
df_junk_food=pd.read_csv('df_item_junk_food.csv')
```

```python
df_junk_food['Main_Cat']=['Junk Food' for i in df_junk_food['Name']]
```

```python
def label_jf(x):
    
    
    if ('巧克力' in x ) or ('朱古力' in x )or ('可可' in x):
        return 'Chocolate'
    elif ('餅' in x )or('沙琪瑪'in x) or ('威化' in x ) or ('百力滋' in x)or ('曲奇'in x) or ('甜蜜時光' in x) or('阿華田' in x):
        return 'Cookies'
    elif ('爆谷' in x ) or('爆米花'in x) :
        return 'Popcorn'
    elif ('芝士圈' in x ) or ('膨化' in x ) or('芝士波'in x)or('芝士條'in x)or('春卷'in x) or('吐司' in x) or ('池田屋' in x) or ('蝦條' in x) or ('脆' in x):
        return 'Crackers'
    elif ('糕' in x ) or ('酥' in x ):
        return 'Cake'
    
    elif ('棒' in x ) or  ('營養麥' in x ):
        return 'snack bar/breakfast'
    elif ('齋'in x)or ('肉乾' in x ) or ('魚絲' in x )or('帆立貝'in x)or ('鱈魚' in x)or ('墨魚'in x) or('魚鬆' in x)or ('螺肉' in x) or ('快車' in x) or ('豬肉紙'in x) or ('肉鬆'in x) or ('魚柳' in x) or ('魷'in x) or('Bento'in x):
        return 'jerky'
    elif ('腸' in x ) or ('魚小食' in x) :
        return 'Sausage'
    
    elif ('西梅' in x ) or ('果乾' in x) or ('八仙果' in x) or ('提子乾' in x) or ('山楂' in x) or ('桃乾' in x) or ('莓乾' in x)or ('元肉'in x):
        return 'Dried Fruit'
    
    elif ('花生' in x ) or ('堅果' in x ) or ('核桃'in x) or ('腰果' in x) or('甘栗'in x)or('瓜子' in x )or ('杏仁' in x) or ('果仁'in x) or ('開心果'in x) or ('豆' in x):
        return 'nut/beans'
    elif ('薯片' in x ) or ('片' in x ) or ('魚皮' in x) or ('薯' in x):
        return 'Chips'
    elif ('海苔' in x ) or ('紫菜' in x )or ('海帶' in x ):
        return 'Seaweed'
    elif ('蛋捲' in x) or ('卷' in x):
        return 'EggRoll'
    elif ('喉糖' in x) or ('軟糖' in x) or ('薄荷糖' in x) or ('薑糖'in x)or ('橡皮糖'in x) or ('糖' in x):
        return 'Sweets'
    
    else:
        return 'Others'


df_junk_food['Sub_Cat']=df_junk_food['Name'].apply(lambda x: label_jf(x))
```
