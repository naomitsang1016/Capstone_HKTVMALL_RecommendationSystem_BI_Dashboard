```python
# import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait

from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains, Chrome
import time
import re
```

```python
df_all=pd.DataFrame()
```

```python
# item_list


# rice


# get the page no.

URL = 'https://www.hktvmall.com/hktv/zh/main/search?page=0&q=%3Areview-avg-rating-desc%3Acategory%3AAA11000000000%3Azone%3Asupermarket%3Astreet%3Amain%3A'
    
driver = webdriver.Chrome()

driver.get(URL) #replacing the requests.get step

subhtml = driver.page_source #replacing the html.text

soup = BeautifulSoup(subhtml, "html.parser")
soup.prettify()


good_num=int(soup.find_all('span', {'class' : 'product-result-count'})[0].text)

if type(good_num/60)!=int:
    page_num=good_num//60+1
else:
    page_num=good_num/60
    
driver.close()

```

```python
check_bug=[]
# all_page

for i in range(page_num)[:]:
    check_bug=[]
    item_code_dict={'item_name':[],'url':[],'reviews':[]}
    URL = 'https://www.hktvmall.com/hktv/zh/main/search?page={}&q=%3Areview-avg-rating-desc%3Acategory%3AAA11000000000%3Azone%3Asupermarket%3Astreet%3Amain%3A'.format(i)
    print('Page:',(i+1),'Left:',(page_num-i-1))
    driver = webdriver.Chrome()

    driver.get(URL) #replacing the requests.get step
    time.sleep(30)
    subhtml = driver.page_source #replacing the html.text

    soup = BeautifulSoup(subhtml, "html.parser")
    soup=soup.find(id="product-result")
    soup.prettify()
    try: 
        check_bug.append(soup.find_all('div', {'class' : 'brand-product-name'})[0].text.strip(')').strip('('))
        for j in range(len(soup.find_all('div', {'class' : 'brand-product-name'}))):
            item_code_dict['item_name'].append(soup.find_all('div', {'class' : 'brand-product-name'})[j].text.strip(')').strip('('))
            try:
                item_code_dict['url'].append(re.search('(?P<url>main/[^\s]+")', re.findall(r'(<a href=".*"?)',str(soup.find_all('div', {'class' : 'product-brief'})[j]))[0]).group("url").replace('"',''))
            except:
                item_code_dict['url'].append(np.nan)
            try:
                item_code_dict['reviews'].append(soup.find_all('span', {'class' : 'review-number'})[j].text.strip(')').strip('('))
            except:
                item_code_dict['reviews'].append(np.nan)



        df=pd.DataFrame(item_code_dict)
        df_all=pd.concat([df_all,df],axis=0)
        df_all.to_csv('supermarket_list.csv',index=False)


        driver.close()
    except:
        driver.close()
        try:
            driver = webdriver.Chrome()

            driver.get(URL) #replacing the requests.get step
            time.sleep(30)
            subhtml = driver.page_source #replacing the html.text

            soup = BeautifulSoup(subhtml, "html.parser")
            soup=soup.find(id="product-result")
            soup.prettify()
            check_bug.append(soup.find_all('div', {'class' : 'brand-product-name'})[0].text.strip(')').strip('('))
            for j in range(len(soup.find_all('div', {'class' : 'brand-product-name'}))):
                item_code_dict['item_name'].append(soup.find_all('div', {'class' : 'brand-product-name'})[j].text.strip(')').strip('('))
                try:
                    item_code_dict['url'].append(re.search('(?P<url>main/[^\s]+")', re.findall(r'(<a href=".*"?)',str(soup.find_all('div', {'class' : 'product-brief'})[j]))[0]).group("url").replace('"',''))
                except:
                    item_code_dict['url'].append(np.nan)
                try:
                    item_code_dict['reviews'].append(soup.find_all('span', {'class' : 'review-number'})[j].text.strip(')').strip('('))
                except:
                    item_code_dict['reviews'].append(np.nan)



            df=pd.DataFrame(item_code_dict)
            df_all=pd.concat([df_all,df],axis=0)
            df_all.to_csv('supermarket_list.csv',index=False)


            driver.close()
        except:
            print('i=',i)
```
