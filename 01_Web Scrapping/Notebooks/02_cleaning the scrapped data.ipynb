## this script cleans the scrapped data (item, url, number of reviews)

```python
import pandas as pd
```

```python
df = pd.read_csv('supermarket_list(junk_food)(uncleaned).csv')
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df.groupby('page').count()

#number of row = last page ---> no missing page
```

```python
df['url'].iloc[4][5:]
```

```python
df['url'] = df['url'].apply(lambda x: 'https://www.hktvmall.com/' + x[5:] if isinstance(x,str) else x)
```

```python
df['url'].iloc[14]
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
# save the cleaned data
df.to_csv('supermarket_list(junk_food)(cleaned).csv')
```

## drop rows with less than 50 reviews

```python
df = df[df['reviews']>49]
```
